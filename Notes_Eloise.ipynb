## Graphique des notes et des matières

```python
#plt.style.use('Solarize_Light2')
#plt.style.use('classic')
plt.style.use('grayscale')
plt.figure(figsize=(16,6))
plt.plot(matiere,df_elo['1er trimestre'], c = 'green', label= u"1er trimestre") 
plt.plot(matiere,df_elo['2ème trimestre'], c = 'red', label= u"2ème trimestre") 
plt.plot(matiere,df_elo['3ème trimestre'], c = 'blue', label= u"3ème trimestre") 
#plt.scatter(matiere,df_elo['1er trimestre'], c = 'green')
#plt.scatter(matiere,df_elo['2ème trimestre'], c = 'red')
#plt.scatter(matiere,df_elo['3ème trimestre'], c = 'blue')
plt.xlabel('Matières') # => libellé axe x
plt.ylabel('Notes sur 20') # => libellé axe y
plt.title(u'Un graphique avec les courbes des notes Eloïse de 4A')  # => titre du graphique
plt.xticks(rotation=45)
plt.grid(True)
plt.legend();
```
